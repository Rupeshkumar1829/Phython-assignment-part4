```python
# Given essay
essay = "  python is a versatile language. it supports object oriented, functional programming. python is widely used in data science.  "

# Step 1: Strip whitespace
clean_essay = essay.strip()
print("Step 1 - Clean Essay:")
print(clean_essay)

# Step 2: Convert to Title Case
title_case = clean_essay.title()
print("\nStep 2 - Title Case:")
print(title_case)

# Step 3: Count "python" (case-insensitive)
count_python = clean_essay.lower().count("python")
print("\nStep 3 - Count of 'python':", count_python)

# Step 4: Replace "python" with "Python &"
replaced_essay = clean_essay.lower().replace("python", "Python &")
print("\nStep 4 - Replaced Essay:")
print(replaced_essay)

# Step 5: Split into sentences
sentences = clean_essay.split(". ")
print("\nStep 5 - Sentences List:")
print(sentences)

# Step 6: Print numbered sentences
print("\nStep 6 - Numbered Sentences:")
for i, sentence in enumerate(sentences, start=1):
    if not sentence.endswith("."):
        sentence += "."
    print(f"{i}. {sentence}")
```
